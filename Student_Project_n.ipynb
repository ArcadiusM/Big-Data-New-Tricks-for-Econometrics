## Classification tree for Titanic data <a name="classification-tree-for-titanic-data"></a>

### Data cleaning and processing <a name="data-cleaning-and-processing"></a>

```python
# Data is from http://biostat.mc.vanderbilt.edu/wiki/Main/DataSets
df = pd.read_csv("data/titanic3.csv")

# Extract variables of interest
df = df[["survived", "pclass", "sex", "age", "sibsp"]]
df.head()
```

```python
# Check shape of data
print("Shape of data frame is:")
print(df.shape)

# Check if there are any missing data
print("\nMissing valus in data frame:")
print(df.isnull().sum())
```

```python
# Remove missing variables as it was done in article
df.dropna(inplace = True)
df.shape
```

Usually before building a model data should be split into train and test data. Training on whole data without using any method such as k-cross validation can result in overfitting. My replication use whole data for training as it was done by the author of an article.

```python
# Create dependent variable and predictors
X_train = df[["pclass", "age"]]
y_train = pd.DataFrame(np.where(df["survived"].values == 1, "lived", "died"), df["survived"].index)

# Check data
y_train.head()
```

### Fitting a decision tree <a name="fitting-a-decision-tree"></a>




Author of an article used in R two methods to build a decision tree first of them was "rpart" function and then prunned decision tree was built. Prunned decision tree means that some parameteres were changed in order to deal with overfitting. In R for an object which we get after using "rpart" to prune it, function "prune" can be used and it was. My replication will be slightly different, because there is no same "prune" function in Python

#### Decision tree without pruning <a name="decision-tree-without-pruning"></a>

```python
# Fit a decision tree using "Gini" criterion for splitting data
# "rpart" resulted with tree of depth 4, so max_depth = 4 was used

clf = DecisionTreeClassifier(max_depth = 4)
clf = clf.fit(X_train,y_train)

# Use export_graphviz to create dot file
graph = Source(export_graphviz(clf, out_file=None, feature_names=["class", "age"], filled = True))

# Display in jupyter notebook
display(SVG(graph.pipe(format='svg')))
```

#### Prunned decision tree <a name="prunned-decision-tree"></a>

Now above decision tree will be prunned. This part is additional, because in R this process was automatically done using a function. Here I use grid search to check which max_depth, max_leaf_nodes, min_samples_leaf, min_samples_split are the best, but at the beginning some assuptions was made.

```python
# Set which parameter will be prunned
param_test1 = {'max_leaf_nodes':range(3,16,1)}

gsearch1 = GridSearchCV(estimator = DecisionTreeClassifier(max_depth = 4),
                        param_grid = param_test1, cv = 3, iid = False )

gsearch1 = gsearch1.fit(X_train,y_train)

# Check best max leaf
gsearch1.best_params_, gsearch1.best_score_
```

```python
# Set which parameter will be prunned
param_test2 = {'min_samples_leaf':range(2,100,1)}

gsearch2 = GridSearchCV(estimator = DecisionTreeClassifier(max_depth = 4, max_leaf_nodes = 7),
                        param_grid = param_test2, cv = 3, iid = False  )

gsearch2 = gsearch2.fit(X_train,y_train)

# Check best max leaf
gsearch2.best_params_, gsearch2.best_score_
```

```python
# Set which parameter will be prunned
param_test3 = {'min_samples_split':range(2,100,1)}

gsearch3 = GridSearchCV(estimator = DecisionTreeClassifier(max_depth = 4, max_leaf_nodes = 7, min_samples_leaf = 6),
                        param_grid = param_test3, cv = 3, iid = False )

gsearch3 = gsearch3.fit(X_train,y_train)

# Check best min_samples_split
gsearch3.best_params_, gsearch3.best_score_
```

```python
# Set which parameter will be prunned
param_test4 = {'max_depth':range(1,5,1)}

gsearch4 = GridSearchCV(estimator = DecisionTreeClassifier(max_leaf_nodes = 7, min_samples_leaf = 6, 
                                                           min_samples_split = 2),
                        param_grid = param_test4, cv = 3, iid = False )

gsearch4.fit(X_train,y_train)

# Check best max_depth
gsearch4.best_params_, gsearch4.best_score_
```

```python
# Fit pruned tree
clf_pruned = DecisionTreeClassifier(max_depth = 4, max_leaf_nodes = 7, min_samples_leaf = 6, min_samples_split = 2)
clf_pruned = clf_pruned.fit(X_train, y_train)

# Use export_graphviz to create dot file
graph_1 = Source(export_graphviz(clf_pruned, out_file=None, feature_names=["class", "age"], filled = True))

# Display in jupyter notebook
display(SVG(graph_1.pipe(format='svg')))
```

### Data partition plot <a name="data-partition-plot"></a>


In order to make decision tree even more understandable it is possible to show this kind of classification problem by using a partition plot, which was built below. Presented plot shows how the tree devides the space into rectangular regions. Unfortunately with below plot we can only show partition of two variables, but decision trees do not have this feature.

```python
# Compute confidence intervals and plot them

# Classify age to different 8 bins
age_bins = pd.cut(df["age"], 8)

# Number of people in every interval
num_in_bins = age_bins.value_counts().sort_index()

# Number of survived people in every interval
num_survived_in_bins = age_bins.where(df["survived"] == 1).value_counts().sort_index()

# Fraction of survived people in every interval
frac_survived = num_survived_in_bins / num_in_bins

# Confidence interval for a binomial proportion. Method wilson used as it was used in an article
intervals = prp.proportion_confint(num_survived_in_bins,num_in_bins, method = "wilson")

# Create list to use in plot function
frac_survived = frac_survived.tolist()

# Create diffrences to use in plot function
difference = intervals[1] - intervals[0]
```

```python
plt.figure(figsize=(9,9))

plt.errorbar(x= np.arange(10, 90, 10), 
             y= frac_survived, 
             yerr = [(diff)/2 for diff in difference],
             fmt='o')
plt.xlabel("Age bin")
plt.ylabel("Fraction survived")
plt.title("Plot survival rates by age bins")
```

<i> "Here we see that survival rates for the youngest passengers were relatively high, and survival rates for older passengers
were relatively low. For passengers between these two extremes, age didn’t matter very much. So what mattered for survival is not so much age, but whether the passenger was a child or elderly." </i>

### Introduction to KNN Classifier <a name="introduction-to-knn-classifier"></a>

K-nearest neighbours (KNN) is one of the classification algorithms not decribed in chosen article. This method is very simple, intuitive and easy to understand. K nearest neighbours is one of the examples of supervised learning algorithms. 

KNN is a non-parametric algorithm which means that there is no assumption about data distribution, so model structure of dataset is not assumed, as it is done for example in linear regression method. Non-parametric algorithms are very useful especially when data is far away from following any mathematical theoretical assumptions which can be helpful in practice.        
In K-nearest neighbours, K is the number of nearest data points. The core deciding factor in this algorithm is the number of neighbors, so mentioned before K. For the simplest case we can use K = 1, but for this number of neighbors there is a high possibility of overfitting data and bad performance in predictions. For K =1 algorithm works as follows: suppose X1 is the point from test data for which label needs to predict. Firstly, the nearest point in training data has to be found and then label of this point is assigned to X1. Procedure works similarly for K closest points, but decision of label is made based on majority vote of all k neighbors. In order to find distance between points can be used any distance measure such as: Euclidean distance, Minkowski distance and Manhattan distance.  

Generally we should use KNN algorithm by following steps:
1. Load the test and training data
2. Choose the number of nearest neighbours - K 
3. For each point in test data:
    + find the distance to all training data points using chosen distance measure
    + store and sort all distances
    + choose the first K neighbours 
    + vote for label for test point by majority of labels present in the chosen training points

### Introduction to ridge regression <a name="introduction-to-ridge-regression"></a>

Ridge regression is very similar to basic simple ordinary least squares. There is only one change in the optimization formula which can improve in comparison to lienear regression especcialy while doing predictions. The ridge regression coefficient estimates are the values that minimize: 

RSS +  $\lambda\sum_{i=1}^{p} (\beta_{i})^2$

Second part: $\lambda\sum_{i=1}^{p} (\beta_{i})^2$ is called shrinkage penalty. The tunning parameter $\lambda$ control the importance of penalty and because of that it has an impact on all regression coefficients estimates. Drawback of this method is that we get biased estimator, but there are also adventages which are related to bias-variance trade-off. Increase in the tunning parameter leads to decreased variance, but increased bias. This method in many cases thanks to shrinkage penalty can perform better predictions than simple least squares.

### Introduction to gradient boosting classifier <a name="introduction-to-gradient-boosting-classifier"></a>

Firstly, it is worth to explain what boosting is. Like bagging mentioned in the article, boosting is an approach which can be applied to many machine learning methods for classification or regression. Bagging uses bootstrap to create multiple datasets for training the method. As a next stage bagging fits a sparate decision tree to each training dataset, and then it combines all decision trees in order to create a single predictive model. Every decision tree is independent to others thanks to using bootstrap to create different training datsets. Boosting method works similarly, but in our case decision trees are grown sequentially. It means that each tree is bulit using information from previously built trees. Boosting method does not involve bootstrap sampling. In this method instead of bootstrap each decision tree is fit on a modified version of the original dataset.

Gradient boosting classifier used here is useful for classification tasks, as the name suggests. In gradient boosting the idea is to take weak learning algorithm or hypothesis and make some corrections that will improve the power of this algorithm/hypothesis. In hypothesis boosting, you check every observation on which statistical learning method was trained on, then you leave the observations which were correctly classified. Then method creates new weak lerner and test it only on the observations that were poorly classified. Next, the examples that were correctly classified are kept.   

The idea described above was used in the AdaBoost (Adaptive Boosting) algorithm. In this algorithm, many weak lerners are created thanks to many decision trees that only have a single split, as it is shown below.

```python
X = df_iris[["SepalLengthCm"]].values
y = df_iris["Species"].values
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state = 123)
# Train and fit decision tree
clf_iris = DecisionTreeClassifier(max_depth = 1, random_state  = 123)
clf_iris = clf_iris.fit(X_train,y_train)

# Use export_graphviz to create dot file
graph_iris = Source(export_graphviz(clf_iris, out_file=None, feature_names=["SepalLengthCm"], filled = True))
# Display in jupyter notebook
display(SVG(graph_iris.pipe(format='svg')))
```

Created instances in the trainig dataset are weighted in the way that larger weghts are assigned to instances which were difficult to classify. To the most difficult training instances weaker learners are added sequentially

Gradient boosting classifiers are the Adaptive Boosting method, but it is combined with weighted minimization. After weighted minimization, all classifiers and weighted inputs are again calculated. the aim of gradient boosting classifiers is to minimize the loss and it operates in the similar way, as gradient descent in a neural network.
